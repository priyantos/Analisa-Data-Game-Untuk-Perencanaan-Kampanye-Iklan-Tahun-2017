***Sekarang Akan Kita Periksa Per Platformnya***

```python
games_data['platform'].value_counts()
```

Dari data diatas bisa kita lihat bahwa PS2 memiliki jumlah penjualan yang sangat banyak. Mari kita teliti lebih dalam distribusinya. Kita akan mencoba melakukan analisa untuk platform yang memiliki jumlah yang sangat banyak berdasarkan value_counts()

```python
# Menghitung total penjualan untuk setiap platform
all_platform_sales = games_data.groupby('platform')['total_sales'].sum().reset_index()
all_platform_sales = all_platform_sales.sort_values(by='total_sales', ascending=False)
```

```python
# Membuat bar plot visualisasi penjualan untuk semua platform
plt.figure(figsize=(12, 6))
sns.barplot(x='total_sales', y='platform', data=all_platform_sales, palette='plasma')
plt.title('Total Penjualan per Platform')
plt.xlabel('Total Penjualan (juta USD)')
plt.ylabel('Platform')
plt.show()
```

Coba sekarang kita lihat penjualan distribusi pada platform terbesar yaitu PS2

```python
# Menemukan platform dengan total penjualan terbesar
big_platform = all_platform_sales[all_platform_sales['total_sales'] == all_platform_sales['total_sales'].max()]
print("Platform dengan total penjualan terbesar :")
print(big_platform)
```

```python
# Memilih data penjualan hanya untuk platform terbesar
data_platform_terbesar = games_data[games_data['platform'] == big_platform['platform'].values[0]]

# Membuat distribusi penjualan per tahun
platform_sales_per_tahun = data_platform_terbesar.groupby('year_of_release')['total_sales'].sum()

# Menampilkan distribusi penjualan
print("Distribusi penjualan per tahun untuk platform terbesar:")
print(platform_sales_per_tahun)
```

```python
# Data distribusi penjualan per tahun
tahun = platform_sales_per_tahun.index
penjualan = platform_sales_per_tahun.values

# Membuat plot
plt.figure(figsize=(10, 6))
plt.plot(tahun, penjualan, marker='o', linestyle='-')
plt.title('Distribusi Penjualan per Tahun untuk Platform Terbesar')
plt.xlabel('Tahun Rilis')
plt.ylabel('Total Penjualan (juta USD)')
plt.grid(True)
plt.show()
```

**Kesimpulan Sementara**

Dari diagram garis (line chart) diatas, tampak platform terbesar pada penjualan ini di rentang tahun 2000 hingga 2016 adalah PS2 dengan total penjualan sekitar 1233.56 juta USD. ## Menguji Hipotesis Rata-rata rating pengguna platform Xbox One dan PC adalah sama dan Rata-rata rating pengguna genre Action dan Sports berbeda

* Bagaimana kita merumuskan hipotesis nol dan hipotesis alternatif
* Berapa tingkat signifikansi yang kita pilih untuk menguji hipotesis

Hipotesis yang kita gunakan:

* Rata-rata rating pengguna platform Xbox One dan PC adalah sama.
* Rata-rata rating pengguna genre Action dan Sports berbeda.

Uji Hipotesis Platform Xbox One dan PC

* H0 : Rata-rata rating pengguna platform Xbox One dan PC adalah SAMA.
* H1 : Rata-rata rating pengguna platform Xbox One dan PC adalah TIDAK SAMA.

Alpha : 0.05

```python
# memeriksa value
games_data_2011['platform'].value_counts()
```

```python
# memfilter kolom yang akan digunakan untuk hipotesis
games_data_2011_pc = games_data_2011[(games_data_2011['platform'] == 'PC') & (games_data_2011['user_score'])]
games_data_2011_xone = games_data_2011[(games_data_2011['platform'] == 'XOne') & (games_data_2011['user_score'])]
```

```python
# Filter data permainan PC tahun 2011 dan hilangkan baris dengan nilai yang hilang pada 'user_score'
games_data_2011_pc = games_data_2011[(games_data_2011['platform'] == 'PC') & games_data_2011['user_score'].notna()]
```

```python
games_data_2011_pc
```

```python
# Filter data permainan Xbox One tahun 2011 dan hilangkan baris dengan nilai yang hilang pada 'user_score'
games_data_2011_xone = games_data_2011[(games_data_2011['platform'] == 'XOne') & games_data_2011['user_score'].notna()]
```

```python
games_data_2011_xone
```

```python
games_data_2011_pc.shape, games_data_2011_xone.shape
```

```python
# mengenralisir varians untuk masing-masing platform dengan konversi persentase
np.var(games_data_2011_pc['user_score'])/100, np.var(games_data_2011_xone['user_score'])/100
```

**Di sini kita akan menggunakan metode uji statistik yang digunakan untuk menguji homogenitas varians antara dua atau lebih kelompok atau sampel dalam analisis statistik.**

```python
# H0 : Rata-rata rating pengguna platform Xbox One dan PC adalah SAMA.
# H1 : Rata-rata rating pengguna platform Xbox One dan PC adalah TIDAK SAMA.
# Alpha : 0.05

t_statistic, p_value = stats.ttest_ind(games_data_2011_pc['user_score'], games_data_2011_xone['user_score'], equal_var=False)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

alpha = 0.05

if p_value < alpha:
    print('MENERIMA H1')
else:
    print('MENERIMA H0')
```

**Kesimpulan Sementara**

* Uji Hipotesis untuk platform Xbox One dan PC: Menerima H0: Rata-rata rating pengguna platform Xbox One dan PC adalah SAMA.

Uji Hipotesis Genre Action dan PC

* H0 : Rata-rata rating pengguna genre Action dan Sports TIDAK SAMA.
* H1 : Rata-rata rating pengguna genre Action dan Sports SAMA.

Alpha : 0.05

```python
# memfilter kolom yang akan digunakan untuk hipotesis
games_data_2011_action = games_data_2011[(games_data_2011['genre'] == 'Action') & (games_data_2011['user_score'])]
games_data_2011_sports = games_data_2011[(games_data_2011['genre'] == 'Sports') & (games_data_2011['user_score'])]
```

```python
games_data_2011_action.shape, games_data_2011_sports.shape
```

```python
# mengenralisir varians untuk masing-masing platform dengan konversi persentase
np.var(games_data_2011_action['user_score'])/100, np.var(games_data_2011_sports['user_score'])/100
```

**Sama seperti sebelumnya, kita akan menggunakan methode Levene**

```python
# H0 : Rata-rata rating pengguna genre Action dan Sports adalah SAMA.
# H1 : Rata-rata rating pengguna genre Action One dan Sports adalah TIDAK SAMA .
# Alpha : 0.05

t_statistic, p_value = stats.ttest_ind(games_data_2011_action['user_score'], games_data_2011_sports['user_score'], equal_var=False)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

alpha = 0.05


if p_value < alpha:
    print('MENERIMA H1')
else:
    print('MENERIMA H0')
```

**Kesimpulan Sementara**

* Uji Hipotesis untuk genre Action dan Sports Menerima H1: Rata-rata rating pengguna genre Action dan Sports adalah TIDAK SAMA.
